## I. Primera parte (50%)

**Pregunta 1.** Considere la siguiente distribución de rendimientos de los activos A, B y C:

| Probabilidad     | Rendimiento A       | Rendimiento B         | Rendimiento C         |
| ---------------- | ------------------  | -------------------   | ------------------    |
| 50%              | 0.10                | 0.03                  | 0.08                  |
| 30%              | 0.20                | 0.15                  | -0.13                 |
| 20%              | -0.15               | -0.10                 | 0.02                  |

¿Cuál es el rendimiento esperado del activo C?

A. 4.00%

B. 5.00%

C. 8.00%

La respuesta correcta es (5%): B

```python
Er=(tabla1['Prob']*tabla1['Rendimiento C']).sum()
Er
```

**Pregunta 3.** Considere la siguiente distribución de rendimientos de los activos A, B y C:

¿Cuál es la el rendimiento esperado y la volatilidad de un portafolio compuesto 40% por el activo A, 40% por el activo B y 20% por el activo C?

A. $E[r_P]$ = 9.40%, $\sigma_P$ = 7.25%

B. $E[r_P]$ = 4.09%, $\sigma_P$ = 7.52%

C. $E[r_P]$ = 4.90%, $\sigma_P$ = 7.52%

D. $E[r_P]$ = 4.90%, $\sigma_P$ = 5.72%

La respuesta correcta es (5%): C

```python
# La justificación a esta pregunta son los cálculos necesarios para llegar al resultado (5%)
tabla3 = pd.DataFrame(data={
    'Prob':[0.50, 0.30, 0.20],
    'A':[0.10, 0.20, -0.15],
    'B':[0.03, 0.15, -0.10],
    'C':[0.08, -0.13, 0.02]
})
tabla3
```

```python
ErA = (tabla3['Prob'] * tabla3['A']).sum()
ErB = (tabla3['Prob'] * tabla3['B']).sum()
ErC = (tabla3['Prob'] * tabla3['C']).sum()
Er = 0.4 * ErA + 0.4 * ErB + 0.2 * ErC
Er
```

```python
sABC = (tabla3['Prob'] * (0.4 * tabla3['A'] + 0.4 * tabla3['B'] + 0.2 * tabla3['C'] - Er)**2).sum()**0.5
sABC
```

**Punto 2 (15%).** Usando los datos que obtuvo para los **activos que seleccionó** en el punto anterior:

1. Construya la frontera de mínima varianza que representa las posibles combinaciones de portafolios que se pueden crear con los activos que **seleccionó en el primer punto**. Reportar los **rendimientos esperados y las volatilidades de estos portafolios (base mensual) en un DataFrame** (8%).

2. Encontrar los pesos del portafolio de mínima varianza. Reportar el **rendimiento esperado mensual y la volatilidad mensual del portafolio de mínima varianza** (7%).

```python
import numpy as np
N = 101
w = np.linspace(0,1,N)

E1 = ret_summary.loc['KO','Mean']
E2 = ret_summary.loc['NKE','Mean']

s1 = ret_summary.loc['KO','Volatility']
s2 = ret_summary.loc['NKE','Volatility']

r12 = cor.loc['NKE','KO']
E1, E2, s1, s2, r12
```

```python
portafolio = pd.DataFrame(data={'w':w,
                                 'Mean':w*E1 + (1-w) * E2,
                                 'Volatility':((w*s1)**2 + ((1-w) * s2)**2 + 2 * w * (1-w) * s1 * s2 * r12)**0.5})
portafolio
```

```python
plt.figure(figsize=(8,6))
plt.plot(portafolio['Volatility'],portafolio['Mean'],'k',lw=2, label = 'Portafolios activos')
plt.plot(s1,E1, 'b*',ms = 20)
plt.plot(s2,E2, 'r*',ms = 20)
plt.text(s1,E1,'KO')
plt.text(s2,E2,'NKE')
plt.xlabel('Volatilidad ($/sigma$)')
plt.ylabel('Rendimiento Esperado ($E[r]$)')
plt.legend(loc='best')
plt.grid()
```

```python
def varianza(w,s1, s2, r12):
    return (w*s1)**2 + ((1-w) * s2)**2 + 2 * w * (1-w) * s1 * s2 * r12
```

```python
s12 = s1 * s2 * r12
w_min = (s2**2 - s12)/(s1**2 + s2**2 - 2*s12)


E_minvar = w_min * E1 + (1-w_min) *E2
s_minvar = varianza(w_min, s1, s2, r12)**0.5

w_min, 1-w_min, E_minvar, s_minvar
```
